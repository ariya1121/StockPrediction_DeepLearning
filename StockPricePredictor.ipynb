```python
Adj_close_price = df[['Adj Close']]
```

```python
max(Adj_close_price.values), min(Adj_close_price.values)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(Adj_close_price)
scaled_data
```

```python
len(scaled_data)
```

```python
x_data = []
y_data = []

for i in range(100, len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])

x_data, y_data = np.array(x_data), np.array(y_data)
```

```python
x_data[0], y_data[0]
```

```python
int(len(x_data)*0.7)
```

```python
2517-100-int(len(x_data)*0.7)
```

```python
splitting_len = int(len(x_data)*0.7)
x_train = x_data[:splitting_len]
y_train = y_data[:splitting_len]

x_test = x_data[splitting_len:]
y_test = y_data[splitting_len:]
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
```

```python
model = Sequential()
model.add(LSTM(128, return_sequences = True, input_shape = (x_train.shape[1],1)))
model.add(LSTM(128, return_sequences = False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
model.fit(x_train, y_train, batch_size = 1, epochs = 2)
```

```python
model.summary()
```

```python
predictions = model.predict(x_test)
```

```python
predictions
```

```python
inv_predictions = scaler.inverse_transform(predictions)
inv_predictions
```

```python
inv_y_test = scaler.inverse_transform(y_test)
inv_y_test
```

```python
rmse = np.sqrt(np.mean( (inv_predictions - inv_y_test)**2))
```

```python
rmse
```

```python
plotting_data = pd.DataFrame(
    {
    'Original_test_data': inv_y_test.reshape(-1),
    'Predictions': inv_predictions.reshape(-1)
    },
    index = df.index[splitting_len+100:]
)
plotting_data.head()
```

```python
def plot_graph(width, height, data, column_name):
    plt.figure(figsize=(width, height))
    data.plot()
    plt.title(column_name, fontsize=20)  # Increase title font size
    plt.xlabel('Years', fontsize=15)      # Increase x-label font size
    plt.ylabel('Test Data', fontsize=15)     # Increase y-label font size
    plt.legend()
    plt.show()

# Call the function to plot the graph
plot_graph(15, 7, plotting_data, 'Test Data')
```

```python
def plot_graph(width, height, data, column_name):
    plt.figure(figsize=(width, height))
    data.plot()
    plt.title(column_name, fontsize=20)  # Increase title font size
    plt.xlabel('Years', fontsize=15)      # Increase x-label font size
    plt.ylabel('Whole Data', fontsize=15)     # Increase y-label font size
    plt.legend()
    plt.show()

# Concatenate the data
whole_data = pd.concat([Adj_close_price[:splitting_len+100], plotting_data], axis=0)

# Plotting the graph
plot_graph(15, 10, whole_data, 'Whole Data of Amazon')
```
